```r
#Instalando biblioteca e dependências
install.packages("GA")
library(GA)
```

```r

f <- function(x) {
  0.960870 - 0.263137*x[1] + 0.444454*x[2] + 0.676087*x[2]^2
}
```

```r
ga_res <- ga(
  type = "real-valued",
  fitness = function(x) -f(x),  # minimização
  lower = c(-sqrt(2), -sqrt(2)),
  upper = c(sqrt(2), sqrt(2)),
  popSize = 50,
  maxiter = 100,
  run = 50
)

summary(ga_res)
```

```r
plot(ga_res)
```

```r

ga_res@solution
ga_res@fitnessValue
```

```r
#visualizando a função em 3D

x1 <- seq(-sqrt(2), sqrt(2), length=50)
x2 <- seq(-sqrt(2), sqrt(2), length=50)

z <- outer(x1, x2, function(a, b) {
  0.960870 - 0.263137*a + 0.444454*b + 0.676087*b^2
})

persp(x1, x2, z,
      theta=30, phi=30,
      col="lightblue",
      xlab="x1", ylab="x2", zlab="f(x)",
      main="Superfície da função")
```
